### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
from matplotlib import pyplot as plt

tiendas = ['tienda', 'tienda2', 'tienda3', 'tienda4']
ingresos = [total_tienda1, total_tienda2, total_tienda3, total_tienda4]

plt.bar(tiendas, ingresos, color='lightgreen')
for i, ingreso in enumerate(ingresos):
    plt.text(i, ingreso, f"${ingreso:,.0f}", ha='center', va='bottom')

plt.ylim(min(ingresos) * 0.8, max(ingresos) * 1.2)

plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.title('Ingresos por Tienda')

plt.show()
```

```python
tiendas = ['tienda', 'tienda2', 'tienda3', 'tienda4']
satisfaccion_promedio = [media_calificacion_tienda1, media_calificacion_tienda2, media_calificacion_tienda3, media_calificacion_tienda4]

plt.barh(tiendas, satisfaccion_promedio, color='yellow')

plt.xlabel('Satisfacción Promedio')
plt.ylabel('Tiendas')
plt.title('Satisfacción Promedio de Clientes por Tienda')

for i, satisfaccion in enumerate(satisfaccion_promedio):
    plt.text(i, satisfaccion, f"{satisfaccion:.3f}", ha='left', va='center')

plt.ylim(-1, 5)
plt.subplots_adjust(left=0.2)
plt.show()
```
